## 3) **Control de flujo** y **bucles**


```python

# if / elif / else
x = 17
if x % 2 == 0:
    print("par")
elif x % 3 == 0:
    print("divisible por 3")
else:
    print("otro")


```

```python
### Accediendo a los elementos de una lista
for i in nums:
    print(i)
```

```python
# for + range 
suma = 0
for i in range(1, 6):  # 1..5
    suma += i
print("Suma 1..5 =", suma)
```

```python
# For loop con enumerate, accedemos al índice y al elemento
frutas = ["manzana", "pera", "banano"]
for idx, f in enumerate(frutas):
    print(idx, f)
```

```python
# list comprehension
impares_cuadrados = [i*i for i in range(1, 16) if i % 2 == 1]
print("Impares^2:", impares_cuadrados)
```


### 4.5 **Indexación simple** y **filtrado simple**
- Columnas: `df["col"]`, `df[["col1","col2"]]`  
- Filtrado básico (una o dos condiciones simples):  
  - `df[df["tienda"] == "A"]`  
  - `df[(df["tienda"] == "A") & (df["mes"] == "2025-02")]`  
- `.loc` (por **etiquetas**) y `.iloc` (por **posición**) para filas/columnas.


```python

# Selección por columnas
ventas = df["ventas"]
ventas.head()
```

```python
## Seleccionando varias columnas por nombre
subset = df[["tienda", "ventas"]]
subset.head()
```

```python
### Creación de columnas.
df_prueba= df.copy()
df_prueba["Número fijo"]=2
df_prueba.head()
```

```python
## Podemos crear columnas a partir de columnas ya existentes?
df_prueba["ventas_cuadrado"]=df_prueba["ventas"]**2
df_prueba.head()
```

```python
# Filtrado básico (condiciones simples)
solo_A = df[df["tienda"] == "A"]
solo_A
```

```python
B_feb = df[(df["tienda"] == "B") & (df["mes"] == "2025-02")]
B_feb
```

```python

# .loc: por etiquetas (filas y columnas)
febrero_cols = df.loc[df["mes"] == "2025-02", ["tienda","mes","ventas"]]
febrero_cols.head()
```

```python
# .iloc: por posición (slicing simple)
primeras_filas = df.iloc[0:5, :]   # primeras 5 filas, todas las columnas
primeras_filas
```

```python
algunas_cols = df.iloc[:, 0:3]     # todas las filas, 3 primeras columnas
algunas_cols
```

```python
casos_especificos = df.iloc[1:4,1:3]
casos_especificos
```
