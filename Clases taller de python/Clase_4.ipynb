# Condicionales

## El condicional (if)

Es quizá la estructura de control más utilizada. A continuación presentamos un sencillo ejemplo para observar cómo es su sintaxis en Python:

```python
if condicion (booleano):
    instruccion
```

````python
x, y = 2, 5
if x != y:
    print(x, "es distinto que", y)
````

## else
Cuando queremos que se haga una instrucción si el condicional es verdadero y otra en caso contrario debemos usar el else (de lo contrario). La sintaxis es:

```python
if condicion (booleano):
    instruccion
else:
    instruccion si la condicion no se cumple
```

````python
if x != y:
    print(x, "es distinto que", y)
else:
    print(x, "y", y, "son iguales")
````

## elif
Cuando queremos que se haga una instrucción dependiendo de multiples condiciones (verificando cada una en cadena) si el condicional es verdadero y otra en caso contrario debemos usar el else (de lo contrario). La sintaxis es:

```python
if condicion1 (booleano):
    instruccion si se cumple condicion1
elif condicion2 (booleano):
    instruccion si se cumple condicion2
else:
    instruccion si no se cumple ninguna condicion anterior
```

````python
x, y = 2, 5
if x < y:
    print(x, "es menor que", y)
elif x > y:
    print(x, "es mayor que", y)
else:
    print(x, "y", y, "son iguales")
````

Algunos de los operadores con los que se pueden hacer comparaciones al momento de usar en `if` son:

| Operador | Resultado
| -------- | ---------
|   ==     | Igualdad
|   !=     | No es igual        
|   <      | Menor que
|   >      | Mayor que
|   <=     | Menor o igual
|   >=     | Mayor o igual
|   not    | Niega una condición
|   in     | Verifica si un elemento está en una lista
|   and    | Verifica si dos condiciones se cumplen
|   or     | Verifica si se cumple al menos una condición
|   is     | Verifica si un objeto es igual a otro

A excepción de `in` y `not`, todos los demás operadores son similares a los usados en otros lenguajes de programación, por lo que no vamos a mirar ejemplos de ellos. Veamos, por tanto, sólo como funcionan los primeros:

````python
x = 2
condicion1 = x == 2
condicion1
````

````python
condicion2 = x == 3
condicion2
````

````python
condicion3 = x < 3
condicion3
````

````python
# Recordemos en breve las tablas lógicas de verdad
print(condicion1 and condicion2) # T & F -> F
print(condicion2 and condicion1) # F & T -> F
print(condicion1 or condicion2) # T | F -> T
print(condicion2 or condicion1) # F | T -> T
````

````python
aleatorio = 4  # después vemos cómo obtener números aleatorios :)
print(aleatorio)

if (aleatorio<8) & (aleatorio>5):
    print(aleatorio, 'es mayor a 5 y menor a 8')

elif (aleatorio>=8):
    print(aleatorio, 'es mayor o igual a 8')

else:
    print(aleatorio, 'es menor o igual a 5.')
````

````python
lado_oscuro = "Anakin Skywalker"
hijo = "Luke Skywalker"

if lado_oscuro == "Anakin Skywalker":
    print("Darth Vader")

if lado_oscuro == hijo:
    print("Luke, no te rindas al odio. Eso lleva al lado oscuro")
````

````python
# Versión "Minimalista"
print(aleatorio, 'es mayor o igual a 8') if aleatorio>=8 else print(aleatorio, 'es menor que 8.')
````

## Loops

Existen dos tipos de loops.
- *while loops* en donde se itera hasta que cierta condición se deje de cumplir
- *for loops* en donde se itera sobre una lista de elementos.


## while

Vamos a mirar la sintaxis del comando `while`, que si bien no es tan usado como los que vamos a ver a continuación, es muy útil porque permite hacer un ciclo mientras se cumpla una condición. Por ejemplo, recorrer los indices de una lista, tupla o cadena menores a un número.


Miremos un par de ejemplos:

**(1)** En este primer ejemplo, imprimimos la lista de todos los elementos de una lista `li`, pero seleccionándolos de `li` por medio de su índice.

````python
li = [3, 6, 9, 11]
i = 0
while i < 4:
    print(li[i])
    i += 1
````

## for

En Python `for` se utiliza para moverse (o iterar) entre los elementos de una secuencia de datos. Su sintaxis es más sencilla que la usada en C o C++, porque en lugar de utilizar un contador cuyo valor va aumentando o disminuyendo durante el ciclo, se toma una secuencia completa (i.e. una lista, una tupla, o una cadena), y se recorren sus elementos en el orden en que aparecen en ella.

Observemos algunos ejemplos:

````python
for x in [3, 9, 12, 4]:
    print(x)
````

````python
prefijos = "JKLMNOPQ"
sufijo = "ack"
for letra in prefijos:
    print(letra + sufijo)
````

`range` es un comando que se utiliza muy a menudo junto a los ciclos `for`, pues sirve para generar una lista con todos los números desde 0 hasta cierto valor. También puede usarse con dos valores, uno como límite inferior y el otro como límite superior, así:

**Nota**:

En Python 3 `range` genera un objeto de tipo `range` en lugar de una lista, por razones de eficiencia. Por ello hay que convertirlo a una lista para observar lo que contiene.

````python
rango = range(10)
````

````python
print(type(rango))
````

````python
print(list(rango))
````

````python
for i in range(10):
    print(i**2)
````

````python
# También se puede iterar sobre un string
for x in "Python":
   print(x)
````

Diferencias con el `for`: El for está ideado para recorrer elementos directamente, mientras que el `while` hace el recorrido mientras se cumpla una condición.

## Controles sobre los *loops*

A veces nos podemos enfrentar a situaciones en las que necesitamos salir completamente de un ciclo si se cumple alguna condición o puede haber pasos del *loop* que nos queremos saltar. Los comandos que permiten este tipo de control sobre los ciclos son:

```python
break, continue, pass
```

Escriba un ciclo que recorra una palabra, imprimiendo cada letra, y se salga del ciclo apenas encuentre la letra `h`

````python
palabra='Cacharro'
for letra in palabra:
    if letra=='h':
        break
    print(letra)
````

Ahora, no detenga el ciclo sino que salte esa letra

````python
palabra='Charro'
for letra in palabra:
    if letra=='h':
        continue
        print('H')
    print(letra)
````

A diferencia de `break` y `continue`, `pass` es una orden que no hace nada y sirve para llenar espacios de código que se requiere por sintaxis.

````python
palabra='Charro'
for letra in palabra:
    if letra=='h': # Dejar sin ninguna instrucción el if me genera un error de sintaxis.

    print(letra)
````

````python
palabra='Charro'
for letra in palabra:
    if letra=='h':
        pass
        print('H')
    print(letra)
````
